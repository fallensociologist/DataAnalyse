# Игры — Формирование модели монетизации. Декомпозиция

**Контекст:** компания «Космические братья» занимается разработкой одноимённой игры для мобильных устройств. 

Как и многие другие, наша компания хочет зарабатывать с помощью рекламы. Однако здесь есть противоречие:

- Пользователь разозлится и уйдёт, если начать показывать ему рекламу раньше, чем игра его затянет.
- Чем позже создатели игры включат рекламу, тем меньше они заработают.

Необходимо помочь бизнесу выбрать оптимальное время для запуска рекламы. Учитывая расходы на продвижение игры, нужно рассчитать её окупаемость при разных сценариях.

**Заказчик:** менеджеры отдела маркетинга нашей компании

**Цель:** сформировать модель монетизации игрового приложения (оптимизация доходов): окупаемость за меньшее раздражение

**Задачи:**
- Провести предобработку (познакомиться с массивом, проверить массив на дубликаты и пропуски) и исследовательский анализ данных
- Проанализировать влияние событий на совершение целевого события
- Проверить статистические гипотезы (например, гипотезу различия времени прохождения уровня между пользователями, которые заканчивают уровень через реализацию проекта, и пользователями, которые заканчивают уровень победой над другим игроком)

По итогам исследовательского анализа необоходимо сделать **презентацию в формате pdf**, а также **подготовить дашборд**.

Состав дашборда должен быть следующим:

1. Диаграмма, отображающая количество событий по периодам
2. Гистограмма, отображающая количество пользователей, пришедших из разных источников
3. Добавлен фильтр дашборда по типу события
