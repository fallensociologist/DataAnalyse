–ì–æ—Ç–æ–≤–æ, –ø–µ—Ä–µ—Ö–æ–¥–∏–º –∫ –ø–æ–∏—Å–∫—É –¥—É–±–ª–∏–∫–∞—Ç–æ–≤

```python
data['locality_name'] = data['locality_name'].str.lower()
len(data['locality_name'].unique())
```

–í–∏–¥–∏–º, —á—Ç–æ –≥–¥–µ-—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —ë, –∞ –≥–¥–µ-—Ç–æ –µ, –∏–∑–±–∞–≤–∏–º—Å—è –æ—Ç —ç—Ç–æ–≥–æ

```python
for i in range(len(data['locality_name'])):
    if '—ë' in data['locality_name'][i]:
        data.loc[i, 'locality_name'] = data['locality_name'][i].replace('—ë','–µ')
data['locality_name'].unique()
```

–¢–µ–ø–µ—Ä—å –ø–æ–ø—Ä–æ–±—É–µ–º –∏–∑–±–∞–≤–∏—Ç—å—Å—è –æ—Ç –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–π –≤–Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤
